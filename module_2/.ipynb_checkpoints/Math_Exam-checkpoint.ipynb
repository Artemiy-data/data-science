```python
# посмотрим как различается распределение оценок студентов в зависимости от качественных признаков
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=stud_math.loc[stud_math.loc[:, column].isin(stud_math.loc[:, column].value_counts().index[:])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
for col in ['school', 'health', 'goout', 'famrel', 'failures', 'Fedu', 'Medu',
       'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
       'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic']:
    get_boxplot(col)
```

По графикам видно, что часть параметров не будут влиять на оценку студентов.
Проверим, есть ли статистическая разница в распределении оценок по номинативным признакам, с помощью теста Стьюдента. 
Проверим нулевую гипотезу о том, что распределения оценок студентов по различным параметрам неразличимы:

```python
def get_stat_dif(column):
    cols = stud_math.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud_math.loc[stud_math.loc[:, column] == comb[0], 'score'],
                     stud_math.loc[stud_math.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
stud_math_categorical.columns
```

```python
for col in ['school', 'health', 'goout', 'famrel', 'failures', 'Fedu', 'Medu',
            'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
            'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
            'higher', 'internet', 'romantic']:
    get_stat_dif(col)
```

Как мы видим, серьёзно отличаются следующие параметры: goout, failures, Fedu, Medu, sex, address, Mjob и schoolsup. Так как Fjob был исключен нашей гипотизой, принимаю решения исключить и Mjob. Итак, в нашем случае важные переменные, которые, возможно, оказывают влияние на оценку, это: age, studytime, freetime, absences, score, goout, failures, Fedu, Medu, sex, address и schoolsup.
